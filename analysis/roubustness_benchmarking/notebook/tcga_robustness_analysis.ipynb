# Sampling Run 

```r
pct <- c(.2,.3,.4,.5,.6,.7,.8,.9) # 5 sampling runs for each pct
```

```r
for(i in pct){
    print(round(i *9082))
}
```

```r
sampling_run <- list()
for(i in c(1:40)){
    sampling_run[[i]]<-readRDS(paste('/mnt/ptemp/arvind/robustness_analysis/TCGA robustness analysis/output/sampling_run_',i,'.rds',sep=""))
}
```

```r
k=1
for(run in sampling_run){
    cat('Running:',k)
    cat('\n')
    obj <- selectX( M = run$M,
                    sample.class = run$sample.class,
                    alteration.class = run$alteration.class,
                    n.cores = 20,
                    min.freq = (0.001*length(run$sample.class)),
                    n.permut = 1000,
                    lambda = 0.3,
                    tao = 1,
                    save.object = FALSE,
                    verbose = FALSE,
                    estimate_pairwise = FALSE,
                    maxFDR = 0.25)

    saveRDS(obj,paste('/mnt/ptemp/arvind/robustness_analysis/TCGA robustness analysis/output_v4/sampling_run_results_',k,'.rds',sep=""))
    cat('\n Running done:',k)
    cat('\n')
    rm(obj)
    k=k+1
}
```

```r
results_list<-list()
for(i in c(1:40)){
    print(i)
    results_list[[i]] <- (readRDS(paste('/mnt/ptemp/arvind/robustness_analysis/TCGA robustness analysis/output_v4/sampling_run_results_',i,'.rds',sep="")))$result
}
```

```r
saveRDS(results_list,file='../output_v4/all_sampling_results.rds')
```

# Analysis of sampling results

```r
full_result <- readRDS('/mnt/ptemp/arvind/catalogue_work_update/data/results/pan_can/pan_can_tcga_run_result.rds')
```

```r
results_list<-readRDS(file='/mnt/ptemp/arvind/robustness_analysis/TCGA robustness analysis/output_v4/all_sampling_results.rds')
```

```r
full_result %>% filter(FDR) %>% count(type)
```

```r
full_data_pairs<-(full_result %>% filter(FDR) )$name
```

```r
total_hits<-list()
co_pct<-list()
me_pct<-list()
tp_pct<-list()
fp_pct<-list()
k=1
for(i in c(1:40)){
    total_hits[[k]]<-(results_list[[i]] %>% filter(FDR) %>% count())$n
    co_pct[[k]]<-(results_list[[i]] %>% filter(FDR &type=='CO') %>% count())$n*100/total_hits[[k]]
    me_pct[[k]]<-100-co_pct[[k]]
    tp_pct[[k]]<-length(intersect(full_data_pairs,(results_list[[i]] %>% filter(FDR))$name))*100/total_hits[[k]]
    fp_pct[[k]]<-100- tp_pct[[k]]
    k=k+1
}
```

```r
total_hits
```

```r
values<-c()
for(i in pct){
    values<-c(values,rep(i,5))
}
```

```r
length(values)
```

```r
df <- data.frame('Sampling_Percentage'= as.factor(values*100))
```

```r
head(df)
```

```r
df$total_hits <- unlist(total_hits)
df$co_pct <- unlist(co_pct)
df$me_pct <- unlist(me_pct)
df$tp_pct <- unlist(tp_pct)
```
