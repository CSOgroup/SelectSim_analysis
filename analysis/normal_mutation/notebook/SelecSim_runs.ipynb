# Paitent Runs

## Normal data run

```r
normal_data <- readRDS('../data/li_et_al_normal_run_data_patient_v2.rds')
```

```r
str(normal_data)
```

```r
normal_result <- selectX(M = normal_data$M,
                    sample.class = normal_data$sample.class,
                    alteration.class = normal_data$alteration.class,
                    n.cores = 15,
                    min.freq = round(0.001*length(normal_data$sample.class)),
                    n.permut = 1000,
                    lambda = 0.3,
                    tao = 1,
                    save.object = FALSE,
                    verbose = FALSE,
                    estimate_pairwise = FALSE,
                    maxFDR = 0.25,
                    seed = 42)
```

```r
normal_result$result
```

```r
saveRDS(normal_result,file='../results/li_et_al_normal_run_result_patient.rds')
```
