```r
filter_run_data <- function(run_data,gene_order,sample_order){
    final_run_data<-run_data
    final_run_data$M$M$missense<-run_data$M$M$missense[gene_order,sample_order]
    final_run_data$M$M$truncating<-run_data$M$M$truncating[gene_order,sample_order]
    final_run_data$alteration.class<- run_data$alteration.class[gene_order]
    final_run_data$sample.class<-run_data$sample.class[sample_order]
    final_run_data$M$tmb$missense<-run_data$M$tmb$missense[sample_order,]
    final_run_data$M$tmb$truncating<-run_data$M$tmb$truncating[sample_order,]
    return(final_run_data)
}
```

```r
samples <- names(tcga_data_blca$sample.class)
```

```r
length(samples)
```

```r
tcga_blca_sampling_run <- list()
sample_blca <- list()
for(i in c(1:10)){
    sample_blca[[i]] <- sample(samples,113)
    tcga_blca_sampling_run[[i]] <- filter_run_data(run_data = tcga_data_blca,
                        gene_order = names(tcga_data_blca$alteration.class),
                        sample_order = sample_blca[[i]])
}
```

### SelectSim Runs

```r
tcga_bladder_result <- list()
```

```r
for(i in c(1:10)){
    cat(paste('Run: ',i,'\n'))
    cancer_result <- selectX(M = tcga_blca_sampling_run[[i]]$M,
                            sample.class = tcga_blca_sampling_run[[i]]$sample.class,
                            alteration.class = tcga_blca_sampling_run[[i]]$alteration.class,
                            n.cores = 15,
                            min.freq = round(0.001*length(tcga_blca_sampling_run[[i]]$sample.class)),
                            n.permut = 1000,
                            lambda = 0.3,
                            tao = 1,
                            save.object = FALSE,
                            verbose = FALSE,
                            estimate_pairwise = FALSE,
                            maxFDR = 0.25,
                            seed = 42)
    saveRDS(cancer_result,file=paste('../results/tcga_blca_sampling_',i,'_run_result_patient.rds',sep=""))
    tcga_bladder_result[[i]]<-cancer_result$result
    rm(cancer_result)    
}
```
